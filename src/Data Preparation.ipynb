# Users Profiling & Impact 

```python
grouped_df = vents_metadata.groupby('user_id')['sentiment_analysis'].mean().reset_index(name='mean_sent')
grouped_df0 = vents_metadata.groupby('user_id')['sentiment_analysis'].median().reset_index(name='median_sent')
grouped_df00 = vents_metadata.groupby('user_id')['sentiment_analysis'].sum().reset_index(name='sum_sent')

grouped_df["median_sent"] = grouped_df0["median_sent"]
grouped_df["sum_sent"] = grouped_df00["sum_sent"]

profile = []
for i in range(len(grouped_df)):
    if grouped_df["mean_sent"][i] > 0.2:
        profile.append("pos")
    elif grouped_df["mean_sent"][i] > -0.2:
        profile.append("neutral")        
    else:
        profile.append("neg")
#print(len(grouped_df))   
#print(len(profile))        
```

```python
del grouped_df["median_sent"]
grouped_df
```

### Sum of reactions

```python
grouped_df2 = vents_metadata.groupby('user_id')['reactions'].sum().reset_index(name='sum_reactions')
grouped_df3 = vents_metadata.groupby('user_id')['reactions'].mean().reset_index(name='mean_reactions')
grouped_df4 = vents_metadata.groupby('user_id')['reactions'].median().reset_index(name='median_reactions')

grouped_df2['mean_reactions'] = grouped_df3['mean_reactions']
grouped_df2['median_reactions'] = grouped_df4['median_reactions']
grouped_df2
```

### Impact users by sum of reactions to post

```python
print("mean value = ",grouped_df2["sum_reactions"].mean())
print("median value = ",grouped_df2["sum_reactions"].median())
print("var value = ",grouped_df2["sum_reactions"].var())
print("max value = ",grouped_df2["sum_reactions"].max())
print("min value = ",grouped_df2["sum_reactions"].min())

profile2 = []
for i in range(len(grouped_df2)):
    if grouped_df2["sum_reactions"][i] > 330:
        profile2.append("top")
    elif grouped_df2["sum_reactions"][i] > 110:
        profile2.append("high")
    elif grouped_df2["sum_reactions"][i] > 15:
        profile2.append("soft")        
    else:
        profile2.append("low")
```

```python
grouped_df["profile"] = profile
grouped_df["impact"] = profile2

grouped_df
```
